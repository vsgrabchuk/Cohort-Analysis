### Когортный анализ

```python
ca.get_cohort_analysis(
    df=df,
    unit_column='customer_unique_id',
    value_column='customer_unique_id',
    time_column='order_approved_at',
    segments_time_interval=['2017-01-01', '2017-12-01'],
    metric='retention'
)
```
